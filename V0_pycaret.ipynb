```python
import pandas as pd

from pycaret.classification import *
from sklearn.preprocessing import *
```

```python
df1=pd.read_csv('./training set values.csv', sep=',')
```

```python
df1.head()
```

```python
df_target = pd.read_csv('training set labels.csv')
```

```python
df2 = pd.merge(df1, df_target)
df2.head()
```

```python
variables_categoricas = [
    'funder',
    'installer',
    'wpt_name',
    'basin',
    'subvillage',
    'region',
    'lga',
    'ward',
    'public_meeting',
    'recorded_by',
    'scheme_management',
    'scheme_name',
    'permit',
    'extraction_type',
    'extraction_type_group',
    'extraction_type_class',
    'management',
    'management_group',
    'payment',
    'payment_type',
    'water_quality',
    'quality_group',
    'quantity',
    'quantity_group',
    'source',
    'source_type',
    'source_class',
    'waterpoint_type',
    'waterpoint_type_group'
]

```

```python
df3 = df2
le = LabelEncoder()
for variable in variables_categoricas:
    df3[variable] = le.fit_transform(df3[variable])
```

```python
df1.head()
```

```python
pd.set_option('display.max_columns', None)
df3.head()
```

```python
model_setup = setup(data=df3, 
                    target = 'status_group', 
                    session_id=1, 
                    )
```

```python
if 1:
    best_model = compare_models()
```

```python
df_target = pd.read_csv('./test set values.csv', sep=',')
```

```python
df_target_enlabeled = df_target.copy()
le = LabelEncoder()
for variable in variables_categoricas:
    df_target_enlabeled[variable] = le.fit_transform(df_target_enlabeled[variable])
```

```python
predictions = predict_model(best_model, data=df_target_enlabeled)
```

```python
predictions
```

```python
df_result = predictions[['id', 'prediction_label']]
```
